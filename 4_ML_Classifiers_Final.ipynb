```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
import pickle

# Load data
df = pd.read_csv("/Users/ismailperacha/Downloads/combined_dataset.csv") #change to your path

# Drop missing labels
df = df.dropna(subset=['label'])

# Separates features and label here
X = df.drop(columns='label')
y = df['label']

# --- Reduced Feature Set via Feature Selection ---
selector = SelectKBest(mutual_info_classif, k=30)
X_selected = selector.fit_transform(X, y)

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Random Forest
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42) #change ML class method here 
rf_clf.fit(X_train, y_train)

# Save model
pickle.dump(rf_clf, open("finalized_model_RF.sav", 'wb'))  #change name of saved model file  

# Load and test
loaded_rf = pickle.load(open("finalized_model_RF.sav", 'rb')) #change loaded_nameofclassifiter

# Evaluate
print("Random Forest Accuracy:", loaded_rf.score(X_test, y_test)) #change _____ Accuracy to aligned ML model
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.naive_bayes import GaussianNB
import pickle

# Loads data
df = pd.read_csv(r"C:\Users\17147\OneDrive\Desktop\ML Project Data\combined_dataset.csv")

# Drops rows with missing labels
df = df.dropna(subset=["label"])

# Separates features and target
X = df.drop(columns="label")
y = df["label"]

# Selects top 30 features
selector = SelectKBest(mutual_info_classif, k=30)
X_selected = selector.fit_transform(X, y)

# Train/Tests split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Trains Naive Bayes model
nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)

# Save model
pickle.dump(nb_clf, open("finalized_model_NB.sav", 'wb'))  #change name of saved model file  

# Load and test
loaded_nb = pickle.load(open("finalized_model_NB.sav", 'rb')) #change loaded_nameofclassifiter

# Evaluation
print("Naive Bayes Accuracy:", nb_clf.score(X_test, y_test))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest, mutual_info_classif
import pickle

# Load data
df = pd.read_csv(r"C:\Users\17147\OneDrive\Desktop\ML Project Data\combined_dataset.csv") 

# Drops rows with missing labels
df = df.dropna(subset=['label'])

# Separates features and target
X = df.drop(columns='label')
y = df['label']

# Feature selection (top 30 here)
selector = SelectKBest(mutual_info_classif, k=30)
X_selected = selector.fit_transform(X, y)

# Splits data
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initializes and trains Decision Tree Classifier
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)

# Save model
pickle.dump(dt_clf, open("finalized_model_DT.sav", 'wb'))  #change name of saved model file  

# Load and test
loaded_dt = pickle.load(open("finalized_model_DT.sav", 'rb')) #change loaded_nameofclassifiter

# Evaluate
print("Decision Tree Accuracy:", loaded_dt.score(X_test, y_test)) #change _____ Accuracy to aligned ML model
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm import SVC
import pickle

# Load data
df = pd.read_csv(r"C:\Users\17147\OneDrive\Desktop\ML Project Data\combined_dataset.csv") #change to your path

# Drop missing labels
df = df.dropna(subset=['label'])

# Separates features and label here
X = df.drop(columns='label')
y = df['label']

# --- Reduced Feature Set via Feature Selection ---
selector = SelectKBest(mutual_info_classif, k=30)
X_selected = selector.fit_transform(X, y)

# split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)


# Train Support Vector Machine (SVM)
svm_clf = SVC(kernel='linear', probability=True)   # enable predict_proba support for evaluation later
print("Starting SVM training...")
svm_clf.fit(X_train, y_train)
print(":white_check_mark: Finished training SVM.")

# Save model
pickle.dump(svm_clf, open("finalized_model_SMV.sav", 'wb'))  #change name of saved model file  

# Load and test
loaded_smv = pickle.load(open("finalized_model_SMV.sav", 'rb')) #change loaded_nameofclassifiter

# Evaluate
print("SVM Accuracy:", loaded_smv.score(X_test, y_test)) #change _____ Accuracy to aligned ML model
```
